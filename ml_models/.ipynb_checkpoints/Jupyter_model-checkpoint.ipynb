```python
import pandas as pd
import numpy as np
import string

from nltk.corpus import stopwords 

import matplotlib.colors as mcolors

import json
from bson import json_util

import gensim
import gensim.corpora as corpora
from gensim.utils import simple_preprocess
from gensim.models import CoherenceModel
from gensim.models import LdaModel
from gensim.utils import lemmatize

#-----------------------------------------------------------------------------------------------------
def getData(filename):
    df = pd.read_json(r'../server/data/MedellinCleaned.json')
    return df
#-----------------------------------------------------------------------------------------------------
def LDAModel(df,numberOfTopics):
    #Remover filas duplicadas
    df = df.drop_duplicates(subset=['respuesta'])

    #remover espacios extra entre palabras
    df['respuesta'] = df['respuesta'].str.replace('  ',' ')
    df['respuesta'] = df['respuesta'].str.strip()
    #transformar a minuscula
    df['respuesta'] = df['respuesta'].str.lower()

    #eliminar puntuacion
    df['respuesta'] = df['respuesta'].str.replace(r'\s+', ' ')
    df['respuesta'] = df['respuesta'].str.replace('[{}]'.format(string.punctuation), '')

    #ods de ods_n como string a n entero
    df.ods = [ int(data[data.find('_')+1:])for data in df.ods ]
    #meta de meta_ods_n como string a n entero
    df.meta = [ data[data.rfind('_')+1:] for data in df.meta ]

    #Tokenize each sentence into a list of words
    def sent_to_words(sentences):
        for sentence in sentences:
            yield(gensim.utils.simple_preprocess(str(sentence)))
    data = df.respuesta.values.tolist()
    data_words = list(sent_to_words(data))
    #Build the bigram and trigram models
    #Bigrams are two words frequently occurring together in the document.
    bigram = gensim.models.Phrases(data_words, min_count=5, threshold=100) # higher threshold fewer phrases.

    #Trigrams are 3 words frequently occurring.
    trigram = gensim.models.Phrases(bigram[data_words], threshold=100)  

    # Faster way to get a sentence clubbed as a trigram/bigram
    bigram_mod = gensim.models.phrases.Phraser(bigram)
    trigram_mod = gensim.models.phrases.Phraser(trigram)
    trigramTry = [trigram_mod[bigram_mod[data]] for data in data_words] #solo para ver
    #stopwords, filtrar palabras comunes como "el,la"
    stop = stopwords.words('spanish')
    stop.extend(['pues','así','acá','hace','ver','vez','si'
    ,'mientras','ahí','cómo','va','aquí','sino'
    ,'ve','demasiado','tambien','mas','problema','problemática','entonces','hacer','mucho','quedo','mismo','momento','pienso','principal','mucha'])
    pattern = r'\b(?:{})\b'.format('|'.join(stop))
    df['respuesta'] = df['respuesta'].str.replace(pattern, '')

    def remove_stopwords(texts):
        return [[word for word in simple_preprocess(str(doc)) if word not in stop] for doc in texts]
    data_words_nostops = remove_stopwords(data_words)
    def make_bigrams(texts):
        return [bigram_mod[doc] for doc in texts]

    def make_trigrams(texts):
        return [trigram_mod[bigram_mod[doc]] for doc in texts]

    data_words_bigrams = make_bigrams(data_words_nostops)
    data_words_trigrams = make_trigrams(data_words_nostops)
    id2word = corpora.Dictionary(data_words_trigrams)
    texts = data_words_trigrams
    corpus = [id2word.doc2bow(text) for text in texts]
    lda_model = gensim.models.ldamodel.LdaModel(
        corpus=corpus,
        id2word=id2word,
        num_topics=numberOfTopics, 
        random_state=100,
        update_every=1,
        chunksize=100,
        passes=10,
        alpha='auto',
        per_word_topics=True)
    return lda_model, corpus, id2word, data_words_trigrams, stop

#-----------------------------------------------------------------------------------------------------
def modelPerplexityCoherenceScore(lda_model,data_words_trigrams):
    # Compute Perplexity
    perplexity = lda_model.log_perplexity(corpus)
    # Compute Coherence Score
    coherence_model_lda = CoherenceModel(model=lda_model, texts=data_words_trigrams, dictionary=id2word, coherence='c_v')
    coherence_lda = coherence_model_lda.get_coherence()
    return perplexity, coherence_lda
#-----------------------------------------------------------------------------------------------------
def storeTopicsData(model, corpus, id2word, stop, numberOfTopics):
    model_info = []
    colors = [color for name, color in mcolors.XKCD_COLORS.items()]
    for topic, words in model.show_topics(formatted=False, num_topics=numberOfTopics):
        words_dict = {}
        for word, importance in words:
            word_obj = {'text': word, 'value': float(importance) * 100}
            words_dict[word] = word_obj
        topic = {
          'words': list(words_dict.values()),
          'color': colors[topic],
        }
        model_info.append(topic)
    json_result = json.dumps(model_info, default=json_util.default)
    with open('../client/src/ModelResults/result.json', "w") as outfile:
        outfile.write(json_result)
        print('{"success":true, "message": "El modelo ha sido procesado correctamente 2"}')
    return model_info
#-----------------------------------------------------------------------------------------------------  

def storeGeneralInsight(df):
    jsonData = {}
    #NUMERO DE RESPUESTAS--------------------------------------------------
    totalRespuestas = len(df.index)
    jsonData['totalRespuestas'] = totalRespuestas
    #RESPUESTAS POR PREGUNTA-----------------------------------------------
    preguntasCount = df['pregunta'].value_counts()
    cantidadDePreguntas = len(preguntasCount)
    preguntasDict = {}
    for preguntaP, cantidadP in preguntasCount.iteritems():
        preguntasDict[preguntaP] = {'pregunta': preguntaP, 'cantidad': cantidadP}
    jsonData['preguntas'] = list(preguntasDict.values())
    #RESPUESTAS POR EDAD-----------------------------------------------
    edades = {}
    for edad, cantidad in df['rangoEdad'].value_counts().iteritems():
        edades[edad] = {'grupo_edades': edad, 'cantidad': cantidad//cantidadDePreguntas}
    jsonData['edad'] = list(edades.values())
    #RESPUESTAS POR SEXO-----------------------------------------------
    sexo = {}
    for genero, cantidad in df['sexo'].value_counts().iteritems():
        sexo[genero] = {'genero': genero, 'cantidad': cantidad//cantidadDePreguntas}
    jsonData['sexo'] = list(sexo.values())
    #RESPUESTAS POR ODS------------------------------------------------
    ods = {}
    for numero, cantidad in df['ods'].value_counts().iteritems():
        ods[numero] = {'ods': numero, 'cantidad': cantidad//cantidadDePreguntas}
    jsonData['porOds'] = list(ods.values())
    #TOP DE PALABRAS----------------------------------------------------
    topPalabras = {}
    for numero, cantidad in df['palabra'].value_counts().iteritems():
        topPalabras[numero] = {'palabra': numero, 'cantidad': cantidad}
    topPalabras = {'topPalabras': list(topPalabras.values())[:10]}
    jsonData['topPalabras'] = list(topPalabras.values())[:10]
    json_result = json.dumps(jsonData, default=json_util.default)
    with open('../client/src/ModelResults/generalResult1.json', "w") as outfile:
        outfile.write(json_result)
        print('{"success":true, "message": "La analitica general"}')
    return jsonData  
```

```python
def plotGraphsPyLDAvis(lda_model, corpus, id2word):
    # Plotting tools
    import pyLDAvis
    import pyLDAvis.gensim  # don't skip this
    import matplotlib.pyplot as plt
    %matplotlib inline
    from wordcloud import WordCloud, STOPWORDS
    import matplotlib.colors as mcolors


    pyLDAvis.enable_notebook()
    vis = pyLDAvis.gensim.prepare(lda_model, corpus, id2word)
    return vis

def plotGraphsWordCloud(lda_model,numberOfTopics,stop):
    import matplotlib.pyplot as plt
    %matplotlib inline
    from wordcloud import WordCloud, STOPWORDS
    import matplotlib.colors as mcolors

    cols = [color for name, color in mcolors.TABLEAU_COLORS.items()]  # more colors: 'mcolors.XKCD_COLORS'

    cloud = WordCloud(stopwords=stop,
                      background_color='white',
                      width=2500,
                      height=1800,
                      max_words=10,
                      colormap='tab10',
                      color_func=lambda *args, **kwargs: cols[i],
                      prefer_horizontal=1.0)

    topics = lda_model.show_topics(formatted=False)

    fig, axes = plt.subplots(numberOfTopics//4,2, figsize=(10,10), sharex=True, sharey=True)

    for i, ax in enumerate(axes.flatten()):
        fig.add_subplot(ax)
        topic_words = dict(topics[i][1])
        cloud.generate_from_frequencies(topic_words, max_font_size=700)
        plt.gca().imshow(cloud)
        plt.gca().set_title('Topic ' + str(i+1), fontdict=dict(size=16))
        plt.gca().axis('off')


    plt.subplots_adjust(wspace=0, hspace=0)
    plt.axis('off')
    plt.margins(x=0, y=0)
    plt.tight_layout()
    plt.show()
    return plt

```

```python
filename = '../server/data/MedellinCleaned.json'
numberOfTopics = 20
df = getData(filename)
lda_model, corpus, id2word, data_words_trigrams, stop = LDAModel(df,numberOfTopics)
storeTopicsData(lda_model, corpus, id2word, stop, numberOfTopics)
storeGeneralInsight(df)
perplexity, coherence_lda = modelPerplexityCoherenceScore(lda_model,data_words_trigrams)
vis = plotGraphsPyLDAvis(lda_model, corpus, id2word)
wordCloud = plotGraphsWordCloud(lda_model,numberOfTopics,stop)
```
